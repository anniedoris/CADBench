# Randomize Material
Now we will load some materials randomly and apply them:

```python
surface_materials_path = './Materials/Surfaces/'
part_materials_path = './Materials/Part/'

# get a list of directory names in the surface materials path
surface_materials = [f for f in os.listdir(surface_materials_path) if os.path.isdir(os.path.join(surface_materials_path, f))]
# get a list of directory names in the part materials path
part_materials = [f for f in os.listdir(part_materials_path) if os.path.isdir(os.path.join(part_materials_path, f))]

# randomly select a surface material
surface_material = AutoPBR(os.path.join(surface_materials_path,np.random.choice(surface_materials)))

# randomly select a part material
part_material = AutoPBR(os.path.join(part_materials_path,np.random.choice(part_materials)))

image, depth = render_shape('test.step', ground_material=surface_material, model_material = part_material, light_intensity=0.5, mesh_uv_scale=4)
```

# Randomize Camera Postion
